```python
import pandas as pd
from allennlp.predictors.predictor import Predictor
from evaluate_by_joining_elements import evaluate_coreference_by_joining_elements
from tqdm import tqdm
```

```python
predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/coref-spanbert-large-2020.02.27.tar.gz")
```

```python
shawshank_result = evaluate_coreference_by_joining_elements('data/annotation/shawshank.script_parsed.txt', 'data/annotation/shawshank.coref.mapped.csv', coreference_model=predictor)
bourne_result = evaluate_coreference_by_joining_elements('data/annotation/bourne.script_parsed.txt', 'data/annotation/bourne.coref.mapped.csv', coreference_model=predictor)
basterds_result = evaluate_coreference_by_joining_elements('data/annotation/basterds.script_parsed.txt', 'data/annotation/basterds.coref.mapped.csv', coreference_model=predictor)
```

```python
n = 0
m = 0
N = 0

for result in [shawshank_result, bourne_result, basterds_result]:
    tags = result['mention_tags']
    for coref_cluster in tqdm(result['gold_clusters'].values()):
        N += len(coref_cluster) - 1
        for i, j in coref_cluster:
            if all(tags[k]=='D' for k in range(i, j + 1)):
                l = i - 1
                while l >= 0 and (tags[l] == 'D' or tags[l] == 'E' or tags[l] == 'X'):
                    l -= 1
                if l >= 0 and tags[l] == 'C':
                    h = l - 1
                    while h >= 0 and tags[h] == 'C':
                        h -= 1
                    if h + 1 >= 0:
                        speaker = (h + 1, l)

                        while h >= 0 and tags[h] != 'D' and tags[h] != 'S':
                            h -= 1
                        if h >= 0 and tags[h] == 'D':
                            while h >= 0 and (tags[h] == 'D' or tags[h] == 'E' or tags[h] == 'X'):
                                h -= 1
                            if h >= 0 and tags[h] == 'C':
                                g = h - 1
                                while g >= 0 and tags[g] == 'C':
                                    g -= 1
                                if g + 1 >= 0:
                                    prev_speaker = (g + 1, h)
                                    
                                    if prev_speaker in coref_cluster:
                                        m += 1
                            
                        if speaker in coref_cluster:
                            n += 1
```
