```python
import collections
import numpy as np
import os
import pandas as pd
import re
import tqdm
```

```python
data_dir = os.path.join(os.getenv("DATA_DIR"), "mica_text_coref/movie_coref/results/coreference/baselines/")
baseline_files = [os.path.join(data_dir, file_) for file_ in os.listdir(data_dir)
                                                    if re.match(r"[\w\.]*lea[\w\.]*tsv$", file_)]
print(len(baseline_files))
```

```python
dfs = []
for file_ in baseline_files:
    df = pd.read_csv(file_, sep="\t")
    dfs.append(df)
df = pd.concat(dfs)
```

```python
df.head()
```

```python
df = df.rename(columns={"metric":"movie", "movie":"metric"})
df.head()
```

```python
df[~df["remove_gold_singletons"] & ~df["provide_gold_mentions"] & (df["movie"] == "all")].sort_values(
    by="F", ascending=False)
```

```python
for column in df.columns[:-3]:
    print(column, collections.Counter(df[column]))
```

```python
df.to_csv(os.path.join(data_dir, "full_length.baseline.tsv"), sep="\t", index=False)
```
