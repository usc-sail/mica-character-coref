```python
data_dict = {}
for dt in movie_data:
    data_dict[dt["movie"]] = {
        "rater": dt["rater"],
        "token": dt["token"],
        "pos": dt["pos"],
        "ner": dt["ner"],
        "parse": dt["parse"],
        "speaker": dt["speaker"],
        "sent_offset": dt["sent_offset"],
        "clusters": dt["clusters"]
    }
with open("/home/sbaruah_usc_edu/mica_text_coref/data/temp/movie_dict.json", "w") as fw:
    json.dump(data_dict, fw, indent=2)
```
