```python
from mica_text_coref.coref.movie_coref.data import CorefCorpus, CorefDocument

import numpy as np
import os
```

```python
corpus = CorefCorpus(
    "/home/sbaruah_usc_edu/mica_text_coref/data/movie_coref/results/regular/"
    "movie.jsonlines")
```

```python
document_lens = [2048, 3072, 4096, 5120]
overlap_lens = [128, 256, 384, 512]

for document_len in document_lens:
    for overlap_len in overlap_lens:
        _document_lens = []
        _overlap_lens = []
        for document in corpus:
            words = document.token
            parse_tags = document.parse
            sentence_offsets = np.array(document.sentence_offsets)
            scene_boundaries = np.zeros(len(words), dtype=int)
            segment_boundaries = np.zeros(len(words), dtype=int)

            found_content_tag = False
            i = 0
            while i < len(words):
                if parse_tags[i] == "S":
                    if found_content_tag:
                        scene_boundaries[i] = 1
                        found_content_tag = False
                    while i < len(words) and parse_tags[i] == "S":
                        i += 1
                else:
                    if parse_tags[i] in "ND":
                        found_content_tag = True
                    i += 1
            
            i = 0
            while i < len(words):
                if parse_tags[i] in "SNC":
                    j = i + 1
                    while j < len(words) and parse_tags[i] == parse_tags[j]:
                        j += 1
                    segment_boundaries[i] = 1
                    i = j
                else:
                    i += 1
            segment_boundaries[0] = 1
            
            print(document.movie)
            i = 0
            while i < len(words):
                j = min(i + document_len, len(words))
                if j < len(words):
                    while j >= i and segment_boundaries[j] == 0:
                        j -= 1
                    k = i + document_len - overlap_len
                    while k >= i and segment_boundaries[k] == 0:
                        k -= 1
                    nexti = k
                else:
                    nexti = j
                if nexti == i:
                    print(f"\t nexti == i!")
                    break
                prefix = " ".join(words[i: i + 50])
                suffix = " ".join(words[j - 50: j])
                si = np.nonzero(sentence_offsets[:,0] == i)[0][0]
                sj = np.nonzero(sentence_offsets[:,1] == j)[0][0]
                print(f"\t words = [{i:5d}, {j:5d}], sentences = [{si:4d}, {sj:4d}],"
                    f" document_len = {j - i:5d}, overlap_len = {j - nexti:5d}")
                if j != len(words):
                    _document_lens.append(j - i)
                    _overlap_lens.append(j - nexti)
                # print(f"\t {prefix}")
                # print(f"\t {suffix}")
                # print()
                i = nexti
            print()
        print(f"document_len = {document_len}, overlap_len = {overlap_len}")
        print(f"average document len = {np.mean(_document_lens):.3f}")
        print(f"average  overlap len = {np.mean(_overlap_lens):.3f}")
        print()
```
