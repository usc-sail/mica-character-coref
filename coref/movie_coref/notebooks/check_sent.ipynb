```python
from mica_text_coref.coref.movie_coref.data import CorefCorpus

import bisect
```

```python
corpus = CorefCorpus("/home/sbaruah_usc_edu/mica_text_coref/data/movie_coref/"
                     "results/regular/movie.jsonlines")
```

```python
n_crossing = 0
crossing_spans = []
for doc in corpus:
    sent_starts = [sent_offset[0] for sent_offset in doc.sentence_offsets]
    for mentions in doc.clusters.values():
        for mention in mentions:
            i = bisect.bisect_right(sent_starts, mention.begin) - 1
            j = bisect.bisect_right(sent_starts, mention.end) - 1
            if i != j:
                n_crossing += 1
                crossing_spans.append([doc, mention, i, j])
```
