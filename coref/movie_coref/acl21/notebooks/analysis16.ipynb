```python
from utils.imdbmovie import get_movies
import pandas as pd
from collections import Counter
```

```python
imdb_ids = index_df.imdb_id.dropna().astype(int).unique()
```

```python
movie_data = get_movies(imdb_ids)
```

```python
genres_col = []
rating_col = []

for _, row in index_df.iterrows():
    genres = None
    rating = None
    try:
        imdb_id = str(int(row.imdb_id)).zfill(7)
        genres = movie_data[imdb_id]["genres"]
    except Exception:
        pass
    try:
        imdb_id = str(int(row.imdb_id)).zfill(7)
        rating = movie_data[imdb_id]["rating"]
    except Exception:
        pass
    genres_col.append(genres)
    rating_col.append(rating)

index_df["genres"] = genres_col
index_df["rating"] = rating_col
```

```python
for _, row in index_df[index_df.parsed].iterrows():
    if pd.notna(row.imdb_id):
        url = f"https://www.imdb.com/title/tt{str(int(row.imdb_id)).zfill(7)}/"
        if row.rating > 8:
            print(row.id, url, row.year, row.title, row.rating, row.genres)
```
